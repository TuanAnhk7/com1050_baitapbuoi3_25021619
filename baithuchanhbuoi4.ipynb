```python
#W4A8
n=int(input())
sum=0
n=str(n)
def reverse(num):
   num=str(num)
   num=num[::-1]
   num=int(num)
   return(num)
while n!=n[::-1]:
    sum=sum+1
    n=int(n)+reverse(n)
    n=str(n)
print(sum,n)
```

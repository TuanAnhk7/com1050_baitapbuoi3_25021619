```python
#W7A5
def count_pairs_with_sum(nums, target):
    count=[]
    for i in range(len(nums)):
        for j in range(i+1,len(nums)):
            if (nums[i]+nums[j])==target:
                count.append([nums[i],nums[j]])
    return len(count)
nums=list(map(int,input().split()))
target=int(input())
print(count_pairs_with_sum(nums, target))
```

```python
#W7A6
def flatten_list(nums):
    lst=[]
    for i in range(1,len(nums)+1):
        if i%2==0:
            for j in nums[i-1][::-1]:
                lst.append(j)
        else:
            for j in nums[i-1]:
                lst.append(j)
    return lst
nums=eval(input())
print(flatten_list(nums))
```

```python
#W7A7
def length_of_lis(nums):
    n=[1]*len(nums)
    for i in range(len(nums)):
        for j in range(len(nums)):
            if nums[i]>nums[j]:
                n[i]=max(n[i],n[j]+1)
    return max(n)
nums=eval(input())
print(length_of_lis(nums))
```

```python
#W7A8
def minInterval(intervals,queries):
    h=[]
    n=[]
    for i in intervals:
        h.append(i[1]-i[0]+1)
    for i in queries:
        ans=[]
        cnt=0
        for j in range(len(intervals)):
            if intervals[j][0]<=i<=intervals[j][1]:
                ans.append(h[j])
                cnt+=1
        if cnt==0:
            ans.append(-1)
        n.append(min(ans))
    return n
intervals=eval(input())
queries=eval(input())
print(minInterval(intervals,queries))
```

```python
#W7A10
def maxSubarraySumMod(nums, m):
    def search_upper_bound(arr, target):
        low, high = 0, len(arr) - 1
        res = -1
        while low <= high:
            mid = (low + high) // 2
            if arr[mid] > target:
                res = mid
                high = mid - 1
            else:
                low = mid + 1
        return res

    def insert_sorted(arr, val):
        low, high = 0, len(arr)
        while low < high:
            mid = (low + high) // 2
            if val < arr[mid]:
                high = mid
            else:
                low = mid + 1
        arr.insert(low, val)

    current_sum = 0
    max_sum = 0
    sorted_prefixes = [0]

    for num in nums:
        current_sum = (current_sum + num) % m
        
        idx = search_upper_bound(sorted_prefixes, current_sum)
        
        if idx != -1:
            max_sum = max(max_sum, (current_sum - sorted_prefixes[idx] + m) % m)
        else:
            max_sum = max(max_sum, current_sum)
            
        insert_sorted(sorted_prefixes, current_sum)
        
    return max_sum
nums=eval(input())
m=int(input())
print(maxSubarraySumMod(nums, m))
```

```python
#W7A11
def prefix(s:str)->list[int]:
    pi=[0]*len(s)
    for i in range(1, len(s)):
        j = pi[i-1]
        while j>0 and s[i]!=s[j]:
            j = pi[j-1]
        if s[i]==s[j]:
            j+=1
        pi[i] = j
    return pi

text = input()
key = input()
exist = [False]*128
for i in text:
    exist[ord(i)] = True
for i in key:
    exist[ord(i)] = True
sepa = '#'
for i in range(128):
    if not exist[i]:
        sepa = chr(i)
        break

pi = prefix(key+sepa+text)
ans = 0
for i in range(len(key)+1, len(pi)):
    if pi[i] == len(key):
        ans+=1

print(ans)
```

```python
#W7A12
def compare(s1:str, s2:str)->bool:
    #s1<s2
    l = min(len(s1), len(s2))
    for i in range(l):
        if ord(s1[i])<ord(s2[i]):
            return True
        elif ord(s1[i])>ord(s2[i]):
            return False
    if len(s1)>len(s2):
        return False
    return True

def merge(a, b):
    ans = []
    pa, pb= 0, 0
    while pa<len(a) and pb<len(b):
        if compare(a[pa][0], b[pb][0]):
            ans.append(a[pa])
            pa+=1
        else:
            ans.append(b[pb])
            pb+=1
    ans.extend(a[pa:])
    ans.extend(b[pb:])
    return ans

def merge_sort(a):
    if len(a)<2:
        return a
    m = len(a)//2
    a1 = a[:m]
    a2 = a[m:]
    a1sort = merge_sort(a1)
    a2sort = merge_sort(a2)
    return merge(a1sort, a2sort)
a = [0]*4
for i in range(4):
    s = input()
    a[i] = (s, chr(65+i))

a = merge_sort(a)

for i in a:
    print(i[1] , end=" ")
```
